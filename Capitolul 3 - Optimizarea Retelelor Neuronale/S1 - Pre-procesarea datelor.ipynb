```python
data = load_iris(return_X_y=True, as_frame=True)
```

# Verificam distributiile variabilelor continue

```python
columns = list(X.columns)
classes = list(y.value_counts().keys())
for column in columns:
    plt.figure()
    for cls in classes:
        sns.distplot(X[column][y==cls], label=str(cls))
        plt.title(column)
```

```python
columns = list(X.columns)
for column in columns:
    plt.figure()
    sns.barplot(X[column])
    plt.title(column)
```

```python
sns.heatmap(X.corr(), annot=True, cmap='Blues')
```
