```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize_activation_function(fcn):
    x_vals = np.linspace(-3, 3, 100)
    
    y_vals = fcn(x_vals)
    
    plt.plot(x_vals, y_vals)
    plt.xlabel("Valoare input")
    plt.ylabel("Valoare output")
    
    plt.title("Vizualizare functie de activare")
```

## Sigmoid

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
visualize_activation_function(sigmoid)
```

## Tanh

```python
def tanh_np(x):
    return np.tanh(x)

def tanh(x):
    return (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
```

```python
visualize_activation_function(tanh)
```

```python
visualize_activation_function(tanh_np)
```

## Tanh vs Sigmoid

```python
plt.figure()
visualize_activation_function(tanh)
visualize_activation_function(sigmoid)
plt.show()
```

## Softmax

```python
def softmax(x):
    return np.exp(x) / np.exp(x).sum()
```

```python
x = np.random.randn(5)
```

```python
plt.bar(np.arange(len(x)),x)
```

```python
plt.bar(np.arange(len(x)), softmax(x))
```

## ReLu

```python
def relu(x):
    return np.maximum(0, x)
```

```python
visualize_activation_function(relu)
```
